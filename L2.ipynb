<b><h1>5. Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.</b></h1>

```python
dataGranada = data[data.Provincia == 'Granada']
dataGranada = dataGranada[['Poblacion','Municipio', 'Provincia']].copy().reset_index()
dataGranada
```

```python
#media
print('media ',dataGranada.Poblacion.mean())
#max
print('max ',dataGranada.Poblacion.max())
#min
print('min ',dataGranada.Poblacion.min())
#desviacion estandar
print('desviacion estandar ',dataGranada.Poblacion.std())
#mediana
print('mediana ',dataGranada.Poblacion.median())
```

<b><h1>6. Realice un histograma con la población de los Municipios para cada una de las provincias.</b></h1>

<b><h1>7. Seleccione al azar cincuenta municipios diferentes de entre los diponibles en el archivo. Asegúrese de que no se repitan. ¿Luego calcule el promedio de la población y la desviación estándar de esto 50 municipios?.</b></h1>

```python
data.Municipio.nunique()
```

```python
dataNoDuplicates = data.drop_duplicates(subset=['Municipio'], keep='first')
dataNoDuplicates = dataNoDuplicates.sample(n=50)
dataNoDuplicates
```

```python
#promedio
print('promedio ',dataNoDuplicates.Poblacion.mean())
#desviacion estandar
print('desviacion estandar ',dataNoDuplicates.Poblacion.std())
```

<b><h1>8. Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.</b></h1>

```python
dataDistancias = data
dataDistancias['Distancia a Madrid'] = numpy.sqrt((dataDistancias['X'] - -3.703797)**2 + (dataDistancias['Y'] - 40.41663)**2)
dataDistancias
```

```python
dataDistancias[dataDistancias.Municipio != 'Madrid'].nlargest(10,'Distancia a Madrid')
```

```python
dataDistancias[dataDistancias.Municipio != 'Madrid'].nsmallest(10,'Distancia a Madrid')
```
